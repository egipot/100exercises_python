```python
#89 Database filter
# Please download the database file (database.db).
# The file contains a table with 50 country names along with their area in square km and population.
# Please use Python to print out those countries that have an area of greater than 2,000,000.

#90 Database to CSV converter
# Export the filtered_rows to a csv file
```

```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("database.db")
```

```python
cur = conn.cursor()
```

```python
cur.execute("SELECT country FROM countries")
```

```python
rows = cur.fetchall()
```

```python
rows
```

```python
print_db= [a for a in cur.execute("SELECT * FROM sqlite_master ")]
print(print_db)  # This will show the available table names
```

```python
# Please use Python to print out those ROWS WHERE countries have an area of greater than 2,000,000.
cur.execute("SELECT * FROM countries WHERE area >= 2000000")
```

```python
filtered_rows = cur.fetchall()
filtered_rows
```

```python
df = pd.DataFrame(filtered_rows, columns=["rank", "country", "area", "population"])
df
```

```python
df.to_csv("90_filtered_rows.csv", index=False)
```
