```python
# 85 Data cleaner
# Create a script that cleans the countries_raw.txt file from unecessary lines and leaves valid country names only

# Open the text file and save as "text_file"
with open ('countries_raw.txt', mode = 'r') as text_file:
  countries_list = text_file.read().split('\n')
```

```python
# Initially, check the data inside the text file. This line is optional
countries_list
```

```python
# To ensure that all items in the original list (countries_list) is iterated, make a copy and use the copied list as iterator.
# This "countries_copy" will separate the iterator versus the resulting list (which we would like to keep as "countries_list")
countries_copy = countries_list.copy()
```

```python
# Filter the unnecessary elements in the resulting list, by iterating using the copied list "countries_copy"
for country in countries_copy:
  if len(country) <= 1 or country == 'Top of Page':
    countries_list.remove(country)
```

```python
# Print the result:
countries_list
```

```python
with open ('countries_list_edited.txt', mode = 'w') as new_text_file:
  for country in countries_list:
    new_text_file.write(country + '\n')
```

```python
#Ardit answer

with open('countries_raw.txt', 'r') as file:
  content = file.readlines()

content = [i.strip('\n') for i in content if '\n' in i]
content = [i for i in content if i != '']
content = [i for i in content if i != 'Top of Page']
content = [i for i in content if len(i) != 1]
```

```python
content
```

```python
with open('countries_clean.txt', 'w') as file:
  for i in content:
    file.write(i + '\n')
```
